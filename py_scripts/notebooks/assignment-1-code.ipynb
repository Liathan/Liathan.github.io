```python
# Import library
import pandas as pd
```

```python
# Load the dataframe
dataframe = pd.read_csv("../../data/df_trees.csv", sep=",", index_col=None)
```

```python
# Check the structure of the dataframe
dataframe.head()
```

```python
# Count the number of occurrencies for each species
counting_species = dataframe['Name'].value_counts(sort=True).reset_index()
```

```python
# Rename the variables
counting_species = counting_species.rename(columns={'index':'Species',
                                                     'Name':'Count'})
```

```python
# Save the dataframe of counting species
counting_species.to_csv("../../data/counting_species.csv", sep=",", index=False)
```
